```python
import json
import numpy as np
import nltk
```

```python


reviews=[]
with open("amavid_reviews.json", 'r') as fin:
    for l in fin:
        reviews.append(json.loads(l))

# print(reviews[50]['reviewText'])
textOnly = [r['reviewText'] for r in reviews]
# print(textOnly[50])

Freqs = [nltk.FreqDist(nltk.tokenize.word_tokenize(text)) for text in textOnly]
allText = " ".join(textOnly)
allText = nltk.tokenize.word_tokenize(allText)
TotalFreq = nltk.FreqDist(allText)

```

```python
with open('n_words.txt', 'r') as fin:
    data = fin.read()
n_words = list(map(lambda x: x.strip(), data.split(",")))
```

```python
n_words
```

```python
bad_revs = np.zeros(len(reviews))
for i in range(len(reviews)):
    bad = 0
    for entry in Freqs[i].items():
        if (entry[0] in n_words):
            bad+= 1
    bad_revs[i] = bad
```

```python
bad_revs = np.asarray(bad_revs).reshape(37126,1)
print(bad_revs.shape)
```

```python
reviews = np.asarray(reviews)

labels = np.asarray([(rate['overall']//3) for rate in reviews]).reshape(37126,)
print(labels.shape)
```

```python
from sklearn.linear_model import LogisticRegression as LR

lr = LR()
lr.fit_transform(bad_revs[:100], labels[:100])
y_pred = lr.predict(bad_revs[:])

print(y_pred)
```
